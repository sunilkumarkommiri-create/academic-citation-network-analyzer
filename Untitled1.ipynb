```python
# ================================
# INSTALL
# ================================
!pip install pymupdf sentence-transformers networkx arxiv matplotlib -q

# ================================
# IMPORTS
# ================================
import fitz
import arxiv
import networkx as nx
import re
import numpy as np
import matplotlib.pyplot as plt
from sentence_transformers import SentenceTransformer

%matplotlib inline

# ================================
# STEP 1: DOWNLOAD PAPERS
# ================================
def download_arxiv_papers(query, max_results=3):
    papers = {}

    search = arxiv.Search(
        query=query,
        max_results=max_results,
        sort_by=arxiv.SortCriterion.Relevance
    )

    for result in search.results():
        title = result.title.replace(" ", "_")[:50]

        print(f"📥 Downloading: {title}")

        pdf_path = f"{title}.pdf"
        result.download_pdf(filename=pdf_path)

        papers[title] = pdf_path

    return papers

topic = "graph neural networks"
paper_files = download_arxiv_papers(topic, max_results=3)

# ================================
# STEP 2: LOAD TEXT
# ================================
def load_pdf_text(file_path):
    doc = fitz.open(file_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

papers = {}
for name, path in paper_files.items():
    papers[name] = load_pdf_text(path)

# ================================
# STEP 3: BETTER CITATION EXTRACTION (FIXED)
# ================================
def extract_citations(text):
    citations = []

    # [1], [2]
    citations += re.findall(r'\[\d+\]', text)

    # (Smith, 2020)
    citations += re.findall(r'\([A-Za-z]+, \d{4}\)', text)

    return citations

# ================================
# STEP 4: BUILD GRAPH
# ================================
G = nx.DiGraph()

for paper, text in papers.items():
    G.add_node(paper)

    cites = extract_citations(text)

    print(f"{paper} → {len(cites)} citations found")  # debug

    for cite in cites:
        G.add_edge(paper, cite)

# ================================
# STEP 5: EMBEDDINGS
# ================================
model = SentenceTransformer("all-MiniLM-L6-v2")

paper_embeddings = {}
for paper, text in papers.items():
    emb = model.encode(text[:2000])
    paper_embeddings[paper] = emb

# ================================
# STEP 6: QUERY
# ================================
def retrieve_relevant_papers(query, top_k=2):
    query_emb = model.encode(query)

    scores = []
    for paper, emb in paper_embeddings.items():
        sim = np.dot(query_emb, emb) / (
            np.linalg.norm(query_emb) * np.linalg.norm(emb)
        )
        scores.append((paper, sim))

    scores.sort(key=lambda x: x[1], reverse=True)
    return scores[:top_k]

query = "most influential methods"
results = retrieve_relevant_papers(query)

print("\n🔎 Relevant Papers:")
for p, score in results:
    print(f"{p} (score: {round(score,3)})")

# ================================
# STEP 7: GRAPH ANALYSIS
# ================================
print("\n📊 Most Influential (by citations):")
for node, deg in sorted(G.in_degree, key=lambda x: x[1], reverse=True):
    print(f"{node}: {deg}")

# ================================
# STEP 8: VISUALIZE GRAPH (FIXED)
# ================================
print("\n🧩 Graph Info:")
print("Nodes:", len(G.nodes()))
print("Edges:", len(G.edges()))

plt.figure(figsize=(8,6))

if len(G.edges()) == 0:
    print("⚠️ No edges found → citation extraction may be weak")
else:
    pos = nx.spring_layout(G, seed=42)

    nx.draw(
        G,
        pos,
        with_labels=True,
        node_size=2500,
        font_size=8
    )

    plt.title("Citation Network")
    plt.show()
```
